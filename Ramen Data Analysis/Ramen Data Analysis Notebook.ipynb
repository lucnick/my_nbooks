```python
import pandas as pd
ramen_data = pd.read_csv('ramen-ratings.csv')

# Displays the first five rows of data
ramen_data.head()
```

```python
# Displays the last 5 rows of data
ramen_data.tail()
```

```python
ramen_data.describe()
```

```python
ramen_data.nunique()
```

```python
# Displays only the rows in which the country is Vietnam
ramen_data.query('Country == "Vietnam"')
```

```python
# Displays only the brand and style columns
ramen_data[['Brand','Style']]
```

```python
# Displays the Country column only
ramen_data[['Country']]
```

```python
# Sorts from highest to lowest rating
ramen_data.sort_values('Stars',ascending=False)
```

```python
# Replaces USA with United States
ramen_data['Country'] = ramen_data.Country.str.replace("USA", "United States")
ramen_data.head()
```

```python
# Number of unique countries
ramen_data['Country'].nunique()
```

```python
# Groups by country, takes the mean, sorts by the average stars, displays first 3 rows.
ramen_data.groupby('Country').mean().sort_values('Stars',ascending=False).head(3)
```

```python
# Same as previous cell, except now from lowest average rating
ramen_data.groupby('Country').mean().sort_values('Stars',ascending=True).head(3)
```

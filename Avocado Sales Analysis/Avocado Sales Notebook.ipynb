```python
# Imports all necessary modules.
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
# Reads the file and displays the first five lines of data.
avocado_data = pd.read_csv('avocado.csv')
avocado_data.head()
```

```python
# Displays info for avocado data
avocado_data.info()
```

```python
# Converting Dtype of "Date" from object to datetime

avocado_data['Date'] = pd.to_datetime(avocado_data['Date'])
avocado_data.info()
```

```python
# Filters for the year 2015, conventional type, and Albany 
avocado_filter = avocado_data.query('year == 2015 and type == "conventional" and region == "Albany"')

# Creates a new dataframe that only includes the intended columns, then displays 5 rows.
avocado_bag = avocado_filter[['Date','Total Bags','Small Bags']]
avocado_bag.head()
```

```python
# Melts the bags columns of the original data

avocado_melted = pd.melt(avocado_data, id_vars='Date',
       value_vars=['Total Bags','Small Bags'],
       var_name='Bags', value_name='Count')

avocado_melted
```

```python
# Plots the melted data.
sns.relplot(data=avocado_melted, kind='line', x='Date', y='Count', hue='Bags')
plt.pyplot.show()
```

```python
# Bin the data into four quantiles: poor, modest, good, excellent.

avocado_data['Sales Volume'] = pd.qcut(avocado_data['Total Volume'], q=4,
        labels=['Poor','Modest','Good','Excellent'])

avocado_data.head()
```

```python
# Display the sales volume counts on an annual basis.
sns.countplot(data=avocado_data, x='year',hue='Sales Volume')
plt.pyplot.show()
```

```python
# Melt the original data to just year and types of bags.
avocado_compare = pd.melt(avocado_data, id_vars='year',
       value_vars=['Small Bags','Large Bags','XLarge Bags'],
       var_name='Bags', value_name='Count')

avocado_compare
```

```python
# Plot the sales of each type of bag.
sns.relplot(data=avocado_compare, kind='line', x='year', y='Count', hue='Bags')
plt.pyplot.show()
```

## Conclusion: Small Bags tend to sell best.
